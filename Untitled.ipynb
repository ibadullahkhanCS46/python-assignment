```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python

df_Salary=pd.read_csv("Economy.csv")
df_Salary.head()
```

```python
df_Salary.describe()
```

```python
plt.title('GDP')
sns.histplot(df_Salary['GDP'])
plt.show()
```

```python
plt.scatter(df_Salary['GDP'], df_Salary['GDP Year-over-Year'], color = 'r')
plt.title('GDP vs GDP Year-over-Year')
plt.xlabel('GDP')
plt.ylabel('GDP Year-over-Year')
#plt.box(False)
plt.show()
```

```python

X = df_Salary[['GDP']]  
y = df_Salary['GDP Year-over-Year']  
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_train = regressor.predict(X_train)
plt.scatter(X_train, y_train, color = 'r')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('GDP')
plt.ylabel('GDP Year-over-Year')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'GDP/GDP Year-over-Year', loc='best', facecolor='green')
plt.box(False)
plt.show()
```

```python
y_pred_train = regressor.predict(X_train)
plt.scatter(X_test, y_test, color = 'r')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('GDP vs GDP Year-over-Year (Test Set)')
plt.xlabel('GDP')
plt.ylabel('GDP Year-over-Year')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'GDP/GDP Year-over-Year', loc='best', facecolor='green')
plt.box(False)
plt.show()
```

```python


print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
